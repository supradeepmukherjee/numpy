```python
import numpy as np
import matplotlib.pyplot as plt
```

# Advance Operation with Business examples

```python
# Data structure: [restaurantId,2021,2022,2023,2024]
salesData=np.array([
    [1,15000,24000,22000,25000],
    [2,16000,25000,23000,26000],
    [3,17000,26000,24000,27000],
    [4,18000,27000,25000,28000],
    [5,19000,28000,26000,29000],
])
print(' ----- Sales Analysis -----')
print(salesData.shape)
print(salesData[:3])
```

### Total Sales/Yr

```python
print(np.sum(salesData[:,1:],0))
```

### Minimum Sales/Restaurant

```python
print(np.min(salesData[:,1:],1))
```

### Maximum Sales/Yr

```python
print(np.max(salesData[:,1:],0))
```

### Avg Sales/Restaurant

```python
print(np.average(salesData[:,1:],1))
```

### Cumulative Sales/Restaurant

```python
print(np.cumsum(salesData[:,1:],1))
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(np.cumsum(salesData[:,1:],1),0))
plt.title('Avg cumulative sales across all restaurants')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.grid()
```

# Vectors

```python
v1=np.array([1,2,3,4,5])
v2=np.array([6,7,8,9,10])
print(v1+v2)
print(v1*v2)
print(np.dot(v1,v2))
print(np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))))
```

### Vectorised Operation

```python
restaurantTypes=np.array(['Bangali','South Indian','North Indian','Chat','Gujarati'])
vectorisedUpper=np.vectorize(str.upper)
print(vectorisedUpper(restaurantTypes)) 
print(np.vectorize(str.upper)(restaurantTypes)) # same as above
```

### Broadcast

```python
monthlyAvg=salesData[:,1:]/12
print(monthlyAvg)
```
